```python
df['log_returns'] = np.log(df['Close'].shift(-1)) - np.log(df['Close'])
```

```python
df.head()
```

We get one empty column because of we shifted the values by one row

```python
df.tail()
```

```python
df.dropna()
```

Now let us do a confidence interval on the returns of this stock

```python
sample_mean = df['log_returns'].mean()
sample_size = df.shape[0] # shape gives the no.of rows and col in an array. thats why we can index to get no. of rows
sample_sd = df['log_returns'].std(ddof=1) / sample_size**0.5 # ddof is degrees of freedom, it is 1 because we are taking sample
```

```python
print(np.std(df['log_returns'])/ sample_size**0.5)
print(df['log_returns'].std(ddof=1) / sample_size**0.5)
```

```python
z_left = norm.ppf(0.05)
z_right = norm.ppf(0.95)
```

Hence, to find the interval we need the mean + Z-value * sd / root(sample_size) [ which is the standard deviation of a sample]

```python
interval_left = sample_mean + z_left*sample_sd
interval_right = sample_mean + z_right*sample_sd
```

```python
# 90% confidence interval tells you that there will be 90% chance that the average stock return lies between "interval_left"
# and "interval_right".

print('90% confidence interval is ', (interval_left, interval_right))
```

```python
df['MA10'] = df['Close'].rolling(10).mean()
```

```python
df['MA50'] = df['Close'].rolling(50).mean()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.info()
```

### Using a strategy of fast and slow signals

if the fast signal (MA10) is higher than the slow signal (MA50), we buy a stock

```python
plt.figure(figsize=(20,8))
plt.plot(df['Date'],df['Close'])
plt.plot(df['Date'],df['MA10'],color='red')  #fast signal
plt.plot(df['Date'],df['MA50'],color='green') #slow signal
plt.xlabel('Date')
plt.ylabel('Close Price')
```

```python
df['Shares'] = [1 if df.loc[ei,'MA10'] > df.loc[ei,'MA50'] else 0 for ei in df.index]
```

```python
df.tail(10)
```

```python
df['Close1'] = df['Close'].shift(-1)
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df['Profit'] = [df.loc[ei,'Close1']-df.loc[ei,'Close'] if df.loc[ei,'Shares'] ==1 else 0 for ei in df.index]
```

```python
df.tail(10)
```

```python
df['Wealth'] = df['Profit'].cumsum()
```

```python
df.tail(10)
```

Let's look at the winnings based on this fast signal-slow signal strategy

```python
df['Wealth'].plot()
plt.title('Total money you win is {}'.format(df.loc[df.index[-2], 'Wealth']))
```
